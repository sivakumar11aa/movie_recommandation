```python
def convert2(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):# here we convret the string type into the list type to work on the data
        if count!=3:
           l.append(i["name"])
           count+=1
        else:
           break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
#crew has many problem
def fetch_dir(obj):
    l=[]
    for i in ast.literal_eval(obj):# here we convret the string type into the list type to work on the data
        if i['job']=='Director':
           l.append(i["name"])
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_dir)
```

```python
def convert3(obj):
    l=[]
    for i in ast.literal_eval(obj):# here we convret the string type into the list type to work on the data
        l.append(i["name"])
    return l
```

```python
movies['keywords']=movies['keywords'].apply(convert3)
```

```python
movies.head(1)
```

```python
#overview colom is in string so we have convert in into the elist
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
#now we have to remove the space between the name to fingd the sam name
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python

movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))  #to stop the reapted word
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#recomand movies based on similarty
def recom(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)

```
